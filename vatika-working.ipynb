```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, TrainingArguments, Trainer
from datasets import Dataset
import torch
import re
import ftfy
```

```python
# Assuming your DataFrame is named 'df' with columns: id, question, context, answer
# Split into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print(f"Training set size: {len(train_df)}")
print(f"Test set size: {len(test_df)}")
```

```python
# Save splits for reference
train_df.to_csv('/kaggle/working/train_split.csv', index=False, encoding='utf-8')
test_df.to_csv('/kaggle/working/test_split.csv', index=False, encoding='utf-8')
print("Saved train and test splits to /kaggle/working/")
```

```python
# Clean text to handle potential encoding issues
def clean_text(text):
    text = ftfy.fix_text(text)  # Fix misencoded text
    text = re.sub(r'[\x00-\x1F\x7F]', '', text)  # Remove control characters
    return text.strip()
```

```python
# Prepare data for QA model
def prepare_qa_data(df):
    qa_data = []
    for _, row in df.iterrows():
        context = clean_text(row['context'])
        answer = clean_text(row['answer'])
        question = clean_text(row['question'])
        answer_start = context.find(answer)
        if answer_start == -1:
            # print(f"Warning: Answer '{answer}' not found in context for id {row['id']}")
            continue
        qa_data.append({
            'id': row['id'],
            'context': context,
            'question': question,
            'answers': {
                'text': [answer],
                'answer_start': [answer_start]
            }
        })
    return qa_data
```

```python
# Prepare train and test data
train_data = prepare_qa_data(train_df)
test_data = prepare_qa_data(test_df)

# Convert to Hugging Face Dataset
train_dataset = Dataset.from_list(train_data)
test_dataset = Dataset.from_list(test_data)
```

```python

# Load tokenizer and model
model_name = "google/muril-base-cased"  # Optimized for Hindi
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
# Tokenize data
def tokenize_function(examples):
    encodings = tokenizer(
        examples['question'],
        examples['context'],
        truncation=True,
        padding='max_length',
        max_length=512,
        return_offsets_mapping=True
    )
    start_positions = []
    end_positions = []
    for i, offset_mapping in enumerate(encodings['offset_mapping']):
        answer_start = examples['answers'][i]['answer_start'][0]
        answer_text = examples['answers'][i]['text'][0]
        answer_end = answer_start + len(answer_text)
        token_start = None
        token_end = None
        for idx, (start, end) in enumerate(offset_mapping):
            if start <= answer_start < end:
                token_start = idx
            if start < answer_end <= end:
                token_end = idx
                break
        if token_start is None or token_end is None:
            token_start = 0
            token_end = 0
        start_positions.append(token_start)
        end_positions.append(token_end)
    encodings['start_positions'] = start_positions
    encodings['end_positions'] = end_positions
    return encodings
```

```python

train_dataset = train_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)
```

```python
# Set training arguments (updated for newer transformers versions)
training_args = TrainingArguments(
    output_dir='/kaggle/working/results',
    eval_strategy='epoch',  # Updated from evaluation_strategy
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    save_strategy='epoch',
    save_total_limit=2,
    report_to="none",
)
```

```python
# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)
```

```python
print("test")
```

```python
# Train the model
try:
    trainer.train()
except Exception as e:
    print(f"Training failed: {e}")
```

```python
import pandas as pd
import re
import ftfy
import torch
import json
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
from datasets import Dataset

# Clean text to handle potential encoding issues
def clean_text(text):
    if not isinstance(text, str):
        return ""  # Handle non-string inputs (e.g., NaN)
    text = ftfy.fix_text(text)  # Fix misencoded text
    text = re.sub(r'[\x00-\x1F\x7F]', '', text)  # Remove control characters
    return text.strip()

# Prepare data for prediction (no answers)
def prepare_qa_data(df):
    qa_data = []
    for _, row in df.iterrows():
        qa_data.append({
            'id': row['id'],
            'question': clean_text(row['question']),
            'context': clean_text(row['context'])
        })
    return qa_data

# Load fine-tuned model and tokenizer
model_path = '/kaggle/working/qa_model'
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForQuestionAnswering.from_pretrained(model_path)
model.eval()

# Assuming df is your input DataFrame with id, question, context
# Example:
# df = pd.DataFrame({
#     'id': ['qa1', 'qa2', 'qa3'],
#     'question': ['भारत की राजधानी क्या है?', 'भारत का सबसे बड़ा शहर कौन सा है?', 'भारत में कितने राज्य हैं?'],
#     'context': ['भारत एक देश है। इसकी राजधानी नई दिल्ली है। सबसे बड़ा शहर मुंबई है। इसमें 28 राज्य हैं।'] * 3
# })

df=pd.read_csv("/kaggle/input/vatika-csv/test.csv")
# Prepare and convert to Hugging Face Dataset
qa_data = prepare_qa_data(df)
dataset = Dataset.from_list(qa_data)
# Tokenize data
def tokenize_function(examples):
    encodings = tokenizer(
        examples['question'],
        examples['context'],
        truncation=True,
        padding='max_length',
        max_length=512,
        return_tensors='pt'
    )
    return {key: val.tolist() for key, val in encodings.items()}

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Generate predictions
predictions = []
with torch.no_grad():
    for example in tokenized_dataset:
        input_ids = torch.tensor([example['input_ids']])
        attention_mask = torch.tensor([example['attention_mask']])
        outputs = model(input_ids, attention_mask=attention_mask)
        start_idx = torch.argmax(outputs.start_logits, dim=1).item()
        end_idx = torch.argmax(outputs.end_logits, dim=1).item()
        tokens = tokenizer.convert_ids_to_tokens(example['input_ids'])
        answer = tokenizer.convert_tokens_to_string(tokens[start_idx:end_idx+1]).replace('##', '').strip()
        predictions.append({
            'id': example['id'],
            'question': example['question'],
            'context': example['context'],
            'predicted_answer': answer
        })

# Group predictions by context
context_to_qas = {}
for pred in predictions:
    context = pred['context']
    if context not in context_to_qas:
        context_to_qas[context] = []
    context_to_qas[context].append({
        'id': pred['id'],
        'question': pred['question'],
        'predicted_answer': pred['predicted_answer']
    })

# Format output as a list of context-qas pairs
output_data = [
    {'context': context, 'qas': qas}
    for context, qas in context_to_qas.items()
]

# Save predictions to JSON
output_path = '/kaggle/working/predictions.json'
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(output_data, f, ensure_ascii=False, indent=2)
print(f"Predictions saved to {output_path}")

# Print first few context-qas pairs for verification
print("First few predictions:")
for item in output_data[:min(2, len(output_data))]:
    print(json.dumps(item, ensure_ascii=False, indent=2))
```
